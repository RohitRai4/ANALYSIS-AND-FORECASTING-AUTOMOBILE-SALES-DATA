#1. what is the average price and average mileage for convertible cars


```python
by_convertible=df.groupby("body-style").get_group("convertible")
by_convertible
```

```python
convertible=by_convertible.groupby(["body-style"])["price"].agg(["mean"])
convertible=convertible.rename(columns={"mean":"Avg-Price"})
convertible
```

```python
convertible=by_convertible.groupby(["body-style"])["average-mileage"].agg(["mean"])
convertible=convertible.rename(columns={"mean":"Avg-Mileage"})
convertible
```

#2. Sort cars by company and price in decending order


```python
df["unique_company"]=pd.DataFrame(df["company"].unique()).rename(columns={0:"unique_company_name"})
un=df.groupby(["unique_company"])["price"].agg(["sum"]).rename(columns={"sum":"Price"})  
un.sort_values(by="Price",ascending=False).astype("int")
```

5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower


```python
rt=df.groupby(["unique_company"])["length"].unique()
```

```python
rt=rt.rename({"length":"Max Length"})
```

```python
de=df.groupby(["unique_company"])["wheel-base"].unique()
se=df.groupby(["unique_company"])["horsepower"].unique()
```

```python
df5=pd.concat([rt,de,se],axis=1)
df6=pd.DataFrame(df5)
df6=df6.rename(columns={"length":"Max Length"})
df6=df6.rename(columns={"wheel-base":"Hig Wheelbase"})
df6=df6.rename(columns={"horsepower":"Max Horsepower"})
df6
```

```python
max_le=df6.agg(["max"])
```

```python
max_le.rename(index={"max":"Porsche"})
```

6. company wise average price of cars and order them in decending order


```python
price=df.groupby(["company"])["price"].agg(["mean"]).sort_values(by="mean",ascending=False)
price.rename(columns={"mean":"Average Price"}).round(0)
```

7) create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"

```python
import numpy as np
```

```python
df["decision"]=np.where((df["wheel-base"]<100) & (df["length"]<175),"Don't Buy","Buy")
```

```python
df.drop("index",axis=1).drop("unique_company",axis=1)
```

8. Select only cars whose decision is "buy"; find average price based on company of car

```python
select=df.groupby(["decision"]).get_group("Buy").drop("index",axis=1).drop("unique_company",axis=1)

select.reset_index().drop("index",axis=1)
select
```

```python
select.groupby(["company"])["price"].agg(["mean"]).rename(columns={"mean":"Avg_Price"}).round(0)
```
